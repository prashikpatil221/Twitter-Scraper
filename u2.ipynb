```python


# Imports
import os
import pandas as pd
```

```python
# Setting variables to be used in format string command below
tweet_count = 5
text_query = "Diwali"
since_date = "2020-06-01"
until_date = "2020-07-31"

# Using OS library to call CLI commands in Python
os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> text-query-tweet.json'.format(tweet_count, since_date, text_query, until_date))
```

```python
# Reads the json generated from the CLI command above and creates a pandas dataframe
tweets_df = pd.read_json('text-query-tweet.json', lines=True)

print(tweets_df.shape)

# Displays first 5 entries from dataframe
tweets_df.head()
```
